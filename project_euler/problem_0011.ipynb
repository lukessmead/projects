```python
# Only need to search down, right, diagonal(down and right), and diagonal(down and left).
# Searching up is is the same as down, left the same as right and diagonal(up and l/r) the same as diagonal(down and r/l)


''' 
Takes a start point in the x and y directions, the direction to move in x and y, and the number of steps to make.
It then calculates the product of the numbers in the array that are found when moving in the way defined by the inputs.

eg:
>>> grid_multiplication(0, 0, 1, 1, 4)
>>> 279496
'''

def grid_multiplication(start_x : int, start_y : int, step_x : int, step_y : int, n_steps : int) -> int:
    '''
    --- Function Description --------------------------------------------------------------------------------------------------
        Takes a start point in the x and y directions, the direction to move in x and y, and the number of steps to make.
        It then calculates the product of the numbers in the array that are found when moving in the way defined by the inputs.
    ---------------------------------------------------------------------------------------------------------------------------
    
    --- Function Inputs -------------------------------------------------------------------------------------------------------
        : int : start_x : The x index to start at.
        : int : start_y : The y index to start at.
        : int : step_x : The distance to step in the x direction. Not absolute, so a step_x of -1 will step backwards.
        : int : step_y : The distance to step in the y direction. Not absolute, so a step_x of -1 will step backwards.
        : int : n_steps : The number of steps in both directions.
    ---------------------------------------------------------------------------------------------------------------------------
    
    --- Function Outputs ------------------------------------------------------------------------------------------------------
        : int : total_product : The total product of the numbers in the grid found by the movement defined by the inputs.
    ---------------------------------------------------------------------------------------------------------------------------
    
    --- Function Examples -----------------------------------------------------------------------------------------------------
        >>> grid_multiplication(0, 0, 1, 1, 4)
        >>> 279496
    ---------------------------------------------------------------------------------------------------------------------------
    '''
    
    # Check types of function inputs:
    if not isinstance(start_x, int): raise ValueError('Please enter an integer > 0 for the start_x argument.')
    if start_x < 0: raise ValueError('Please enter an integer >= 0 for the start_x argument.')
    if not isinstance(start_y, int): raise ValueError('Please enter an integer > 0 for the start_y argument.')
    if start_y < 0: raise ValueError('Please enter an integer >= 0 for the start_y argument.')
    if not isinstance(step_x, int): raise ValueError('Please enter an integer > 0 for the step_x argument.')
    if not isinstance(step_y, int): raise ValueError('Please enter an integer > 0 for the step_y argument.')
    if not isinstance(n_steps, int): raise ValueError('Please enter an integer > 0 for the n_steps argument.')
    if n_steps < 0: raise ValueError('Please enter an integer >= 0 for the n_steps argument.')
    
    total_product = 1
    
    # Step through the grid indices as defined by the function inputs.
    for step in range(n_steps):
        x = start_x + (step * step_x)
        y = start_y + (step * step_y)
        
        # Check x and y are valid
        if y > len(number_grid) - 1:
            return -1
        if x > len(number_grid[y]) - 1:
            return -1
        
        # Find the number in the grid
        number_in_grid = number_grid[y][x]
        
        # Multiply it to the total product.
        total_product *= number_in_grid
    
    return total_product
```

```python
''' 
Returns the numbers used in the grid multiplication (for use in stating the solution to the problem).

eg:
>>> grid_retrieval(0, 0, 1, 1, 4)
>>> ['8', '49', '31', '23']
'''

def grid_retrieval(start_x : int, start_y : int, step_x : int, step_y : int, n_steps : int) -> list:
    '''
    --- Function Description --------------------------------------------------------------------------------------------------
        Takes a start point in the x and y directions, the direction to move in x and y, and the number of steps to make.
        It then calculates the product of the numbers in the array that are found when moving in the way defined by the inputs.
    ---------------------------------------------------------------------------------------------------------------------------
    
    --- Function Inputs -------------------------------------------------------------------------------------------------------
        : int : start_x : The x index to start at.
        : int : start_y : The y index to start at.
        : int : step_x : The distance to step in the x direction. Not absolute, so a step_x of -1 will step backwards.
        : int : step_y : The distance to step in the y direction. Not absolute, so a step_x of -1 will step backwards.
        : int : n_steps : The number of steps in both directions.
    ---------------------------------------------------------------------------------------------------------------------------
    
    --- Function Outputs ------------------------------------------------------------------------------------------------------
        : list : numbers_in_grid : The numbers in the grid used in the grid multiplication.
    ---------------------------------------------------------------------------------------------------------------------------
    
    --- Function Examples -----------------------------------------------------------------------------------------------------
        >>> grid_retrieval(0, 0, 1, 1, 4)
        >>> ['8', '49', '31', '23']
    ---------------------------------------------------------------------------------------------------------------------------
    '''
    
    # Check types of function inputs:
    if not isinstance(start_x, int): raise ValueError('Please enter an integer > 0 for the start_x argument.')
    if start_x < 0: raise ValueError('Please enter an integer >= 0 for the start_x argument.')
    if not isinstance(start_y, int): raise ValueError('Please enter an integer > 0 for the start_y argument.')
    if start_y < 0: raise ValueError('Please enter an integer >= 0 for the start_y argument.')
    if not isinstance(step_x, int): raise ValueError('Please enter an integer > 0 for the step_x argument.')
    if not isinstance(step_y, int): raise ValueError('Please enter an integer > 0 for the step_y argument.')
    if not isinstance(n_steps, int): raise ValueError('Please enter an integer > 0 for the n_steps argument.')
    if n_steps < 0: raise ValueError('Please enter an integer >= 0 for the n_steps argument.')
    
    numbers_in_grid = []
    
    # Step through the grid indices as defined by the function inputs.
    for step in range(n_steps):
        x = start_x + (step * step_x)
        y = start_y + (step * step_y)
        
        # Check x and y are valid
        if y > len(number_grid) - 1:
            return []
        if x > len(number_grid[y]) - 1:
            return []
        
        # Find the number in the grid
        number_in_grid = number_grid[y][x]
        
        # Multiply it to the total product.
        numbers_in_grid.append(str(number_in_grid))
    
    return numbers_in_grid
```

```python
''' 
Loops through each position in the number grid and tests if it can search each of the four directions.
If it can, it performs the grid multiplication and compares it to the current greatest product, if it is higher
then the new greatest product is stored as well as additional information.

eg:
>>> grid_search(3)
>>> (811502, 6, 12, 'diagonal(down and left)')
'''

def grid_search(search_distance : int) -> list:
    '''
    --- Function Description --------------------------------------------------------------------------------------------------
        Loops through each position in the number grid and tests if it can search each of the four directions.
        If it can, it performs the grid multiplication and compares it to the current greatest product, if it is higher
        then the new greatest product is stored as well as additional information.
    ---------------------------------------------------------------------------------------------------------------------------
    
    --- Function Inputs -------------------------------------------------------------------------------------------------------
        : int : search_distance : The length of the product in the grid.
    ---------------------------------------------------------------------------------------------------------------------------
    
    --- Function Outputs ------------------------------------------------------------------------------------------------------
        : list : Contains the following information:
            : int : greatest_product : The largest product found in the grid.
            : int : greatest_product_x : The starting x index for the product.
            : int : greatest_product_y : The starting y index for the product.
            : str : greatest_product_direction : The direction from the starting x and y to move.
    ---------------------------------------------------------------------------------------------------------------------------
    
    --- Function Examples -----------------------------------------------------------------------------------------------------
        >>> grid_search(3)
        >>> (811502, 6, 12, 'diagonal(down and left)')
    ---------------------------------------------------------------------------------------------------------------------------
    '''
    
    # Check types of function inputs:
    if not isinstance(search_distance, int): raise ValueError('Please enter an integer > 0 for the search_distance argument.')
    if search_distance < 0: raise ValueError('Please enter an integer >= 0 for the search_distance argument.')
    
    greatest_product = -1
    greatest_product_x = -1
    greatest_product_y = -1 
    greatest_product_direction = ''
    
    for current_y in range(len(number_grid)):
        
        for current_x in range(len(number_grid[current_y])):
            
            # Search down
            if (current_y + search_distance < len(number_grid)):
                test_product = grid_multiplication(current_x, current_y, 0, 1, search_distance)
                if test_product > greatest_product: 
                    greatest_product = test_product
                    greatest_product_x = current_x
                    greatest_product_y = current_y
                    greatest_product_direction = 'down'
            
            # Search right
            if (current_x + search_distance < len(number_grid[current_y])):
                test_product = grid_multiplication(current_x, current_y, 1, 0, search_distance)
                if test_product > greatest_product: 
                    greatest_product = test_product
                    greatest_product_x = current_x
                    greatest_product_y = current_y
                    greatest_product_direction = 'right'
            
            # Search diagonal(down and right)
            if (current_y + search_distance < len(number_grid)) and (current_x + search_distance < len(number_grid[current_y])):
                test_product = grid_multiplication(current_x, current_y, 1, 1, search_distance)
                if test_product > greatest_product: 
                    greatest_product = test_product
                    greatest_product_x = current_x
                    greatest_product_y = current_y
                    greatest_product_direction = 'diagonal(down and right)'
            
            # Search diagonal(down and left)
            if (current_y + search_distance < len(number_grid)) and (current_x - search_distance >= -1):
                test_product = grid_multiplication(current_x, current_y, -1, 1, search_distance)
                if test_product > greatest_product: 
                    greatest_product = test_product
                    greatest_product_x = current_x
                    greatest_product_y = current_y
                    greatest_product_direction = 'diagonal(down and left)'
            
    return greatest_product, greatest_product_x, greatest_product_y, greatest_product_direction
```

```python
product_chain_length = 4

solution = grid_search(product_chain_length)

if solution[3] == 'down':
    solution_numbers = grid_retrieval(solution[1], solution[2], 0, 1, product_chain_length)
if solution[3] == 'right':
    solution_numbers = grid_retrieval(solution[1], solution[2], 1, 0, product_chain_length)
if solution[3] == 'diagonal(down and right)':
    solution_numbers = grid_retrieval(solution[1], solution[2], 1, 1, product_chain_length)
if solution[3] == 'diagonal(down and left)':
    solution_numbers = grid_retrieval(solution[1], solution[2], -1, 1, product_chain_length)

print(f'The greatest product of {product_chain_length} adjacent numbers in the same direction is:')
print(' * '.join(solution_numbers) + f' = {solution[0]:,}')
print(f'which are the {product_chain_length} numbers starting at positon ({solution[1]}, {solution[2]}) and moving {solution[3]}.')
```
